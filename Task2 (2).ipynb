```python
def date_type(df):
    '''change the date column in a data frame to datetime'''
    df = df.copy()
    df['time']= pd.to_datetime(df['time'])
    return df
```

```python
x_train=date_type(x_train)
x_test=date_type(x_test)
y_train=date_type(y_train)
```

```python
def extract_datetime_features(df):
    '''Extracting some datetime features
    like year, month, day of month, and day of week'''
    df = df.copy()
    df['year'] = df['time'].dt.year
    df['month'] = df['time'].dt.month
    df['seconds_of_day'] = df['time'].dt.hour * 3600 + df['time'].dt.minute * 60 + df['time'].dt.second
    return df
```

```python
x_train= extract_datetime_features(x_train)
x_test= extract_datetime_features(x_test)
y_train= extract_datetime_features(y_train)
```

```python
for i in data:
    Q1 = data[i].quantile(0.25)
    Q3 = data[i].quantile(0.75)
    IQR = Q3-Q1
    up = Q3 + 1.5*IQR
    low = Q1 - 1.5*IQR

    if data[(data[i] > up) | (data[i] < low)].any(axis=None):
        print(i,"yes")
    else:
        print(i, "no")
```

```python
#removing outlier except for the last column
for i in data.iloc[:,:-1]:
    Q1 = data[i].quantile(0.25)
    Q3 = data[i].quantile(0.75)
    IQR = Q3 - Q1
    up_lim = Q3 + 1.5 * IQR
    low_lim = Q1 - 1.5 * IQR
    data.loc[data[i] > up_lim,i] = up_lim
    data.loc[data[i] < low_lim,i] = low_lim
```
